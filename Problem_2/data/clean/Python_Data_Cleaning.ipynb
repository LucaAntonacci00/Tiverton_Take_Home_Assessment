## Exploratory Data Analysis + Cleaning

```python
import pandas as pd
import numpy as np


df = pd.read_csv("crop_insurance_deals_2023_Problem2.csv")
df.head()

```

### Standardize Column Names

```python
df.columns = [c.strip().upper() for c in df.columns]
df.columns
```

### Drop Duplicates

```python
dup = df.duplicated().sum()
print("Duplicate full rows:", dup)

if dup > 0:
    df = df.drop_duplicates()
    print("After dropping duplicates:", df.shape)
```

### Convert Numeric columns

```python
import numpy as np

expected_numeric = [
    "CROP_YEAR",
    "STATE_CODE", "COUNTY_CODE", "CROP_CODE",
    "COVERAGE_LEVEL",
    "POLICIES_SOLD", "POLICIES_INDEMNIFIED",
    "LIABILITY", "TOTAL_PREMIUM", "INDEMNITY",
    "LOSS_RATIO"
]
numeric_cols = [c for c in expected_numeric if c in df.columns]
print("Numeric cols found:", numeric_cols)
```

```python
for col in numeric_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("$", "", regex=False)
        .str.strip()
        .replace({"": np.nan, "nan": np.nan, "None": np.nan})
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[numeric_cols].dtypes
```

### Policies indemnified cannot exceed policies sold

```python
if "POLICIES_SOLD" in df.columns and "POLICIES_INDEMNIFIED" in df.columns:
    bad = (df["POLICIES_INDEMNIFIED"] > df["POLICIES_SOLD"]).sum()
    print("Rows where indemnified > sold:", bad)

    df = df[df["POLICIES_INDEMNIFIED"] <= df["POLICIES_SOLD"]]
    print("After removing invalid policy rows:", df.shape)
```

### No Negative Dollars

```python
money_cols = [c for c in ["LIABILITY", "TOTAL_PREMIUM", "INDEMNITY"] if c in df.columns]

for c in money_cols:
    print(c, "negative rows:", (df[c] < 0).sum())

for c in money_cols:
    df = df[df[c] >= 0]

print("After removing negative financial rows:", df.shape)
```

### Recompute Loss Ratio

```python
if "INDEMNITY" in df.columns and "TOTAL_PREMIUM" in df.columns:
    df["LOSS_RATIO_CALC"] = np.where(
        df["TOTAL_PREMIUM"] > 0,
        df["INDEMNITY"] / df["TOTAL_PREMIUM"],
        np.nan
    )

    if "LOSS_RATIO" in df.columns:
        diff = (df["LOSS_RATIO"] - df["LOSS_RATIO_CALC"]).abs()
        print("Loss ratio difference summary (raw vs calc):")
        print(diff.describe())

    df["LOSS_RATIO"] = df["LOSS_RATIO_CALC"]
    df.drop(columns=["LOSS_RATIO_CALC"], inplace=True)
```

### Any Missing Critical Fields?

```python
critical = [c for c in ["STATE_ABBR", "CROP_NAME", "POLICIES_SOLD", "POLICIES_INDEMNIFIED", "LIABILITY", "TOTAL_PREMIUM", "INDEMNITY"] if c in df.columns]
df[critical].isna().sum().sort_values(ascending=False)
```

### Drop Rows missing financial info

```python
df = df.dropna(subset=[c for c in ["LIABILITY", "TOTAL_PREMIUM", "INDEMNITY"] if c in df.columns])
print("After dropping rows missing money fields:", df.shape)
```

```python
print("Final row count:", len(df))

money_cols = [c for c in ["LIABILITY", "TOTAL_PREMIUM", "INDEMNITY"] if c in df.columns]
print("\nMissing in key money fields:")
print(df[money_cols].isna().sum())

print("\nLoss ratio range:")
if "LOSS_RATIO" in df.columns:
    print(df["LOSS_RATIO"].describe())
```

### Cleaning white space/naming conventions

```python
text_cols = ["STATE_ABBR", "COUNTY_NAME", "CROP_NAME"]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
```

```python
text_cols = ["STATE_ABBR", "COUNTY_NAME", "CROP_NAME"]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
```

```python
# Clean crop names thoroughly
df["CROP_NAME"] = (
    df["CROP_NAME"]
    .astype(str)
    .str.strip()                 
    .str.replace(r"\s+", " ", regex=True)  
    .str.lower()                 
    .str.title()                 
)

df["CROP_NAME"].value_counts()
```
